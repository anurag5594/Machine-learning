```python
# Bubble graph on a sample data for a reference

import matplotlib.pyplot as plt

# Sample Data
age = [25, 30, 35, 40, 45, 50, 55, 60]
income = [45, 55, 65, 75, 85, 95, 105, 115]  # Annual Income in k$
spending_score = [60, 65, 70, 80, 85, 90, 50, 40]
gender = ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']

# Bubble size based on Annual Income (scaled down for visualization)
bubble_size = [i * 10 for i in income]

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(age, spending_score, s=bubble_size, c=income, cmap='viridis', alpha=0.6, edgecolors="w", linewidth=2)

# Labels
plt.title("Bubble Plot: Age vs Spending Score vs Annual Income")
plt.xlabel("Age")
plt.ylabel("Spending Score")

# Show color bar for income
plt.colorbar(label="Annual Income (k$)")

plt.show()
```
