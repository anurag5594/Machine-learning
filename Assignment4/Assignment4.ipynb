```python
import pandas as pd
```

```python
# reading the uploaded file

file = pd.read_csv("iris.csv")
```

```python
file.head()
```

```python
file.shape
```

```python
X = file.drop(columns=["variety"]) # declaring the data for X-axis
y = file["variety"] # declaring the data for y-axis
```

```python
from sklearn.preprocessing import LabelEncoder

# given variety has three types "setosa", "virginica", "versicolor"
le = LabelEncoder()
y = le.fit_transform(y) #lableEncoder converts the values to numeric values like 0,1,2
```

```python
from sklearn.model_selection import train_test_split
# creating train set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

# using a decision tree to classify
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score as cm
```

```python
cm = confusion_matrix(y_test, y_pred)
# plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

plt.show()

```

```python
# getting a classification report on Accuracy, Recall, F1, Precision

from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy : " , accuracy)

precision = precision_score(y_test, y_pred, average='weighted')
print("Precision : " , precision)

recall = recall_score(y_test, y_pred, average='weighted')
print("Recall : " , recall)

f1 = f1_score(y_test, y_pred, average='weighted')
print("F1 Score : " , f1)
```
